## StateVector

[Documentation](https://docs.quantum.ibm.com/api/qiskit/qiskit.quantum_info.Statevector)

A single qubit could be constructed as follows:

```python
from qiskit.quantum_info import Statevector

sv = Statevector.from_label('0')

print(sv)
```

The above code generates a single qubit in state $\vert0\rangle$. It prints out as `Statevector([1.+0.j, 0.+0.j], dims=(2,))`.

The first element indicates the probability and relative phase for $\vert0\rangle$. Here the probablity of state $\vert0\rangle$ is 1 and the phase $ψ$ is 0. Note that the data array is of [complex datatype](https://docs.python.org/3/library/functions.html#complex).

Similarly, we could generate $\vert1\rangle$, $\vert+\rangle$, $\vert-\rangle$, $\vert+i\rangle$, $\vert-i\rangle$

```python
sv_1 = Statevector.from_label('1')
sv_plus = Statevector.from_label('+')
sv_minus = Statevector.from_label('-')
sv_plus_i = Statevector.from_label('r')
sv_minus_i = Statevector.from_label('l')

print(sv_1)
print(sv_plus)
print(sv_minus)
print(sv_plus_i)
print(sv_minus_i)
```

$\vert-\rangle$ is equivilant to $$\vert-\rangle = \frac{1}{\sqrt{2}}(\vert0\rangle-\vert1\rangle)$$

This matches then output `Statevector([ 0.70710678+0.j, -0.70710678+0.j], dims=(2,))`.

Milti-quibits could be defined as follows:

```python
sv_multi = Statevector.from_label('001001')

print(sv_multi)
```

The above example defines a qubit in state $\vert001001\rangle$. The output of the state vector contains $2^n$ element and $i^{th}$ element indicating the probablity and phase of the binary representation of the state in $i$.

In the exaple above, only the $9th$ element is `1.+0.j`, which is exactly binary `001001`.

Note: The qubit is started from right to left. That being said, firsit quibit (qubit 0) is $\vert1\rangle$ and the second qubit is $\vert0\rangle$

## Circuits

A very simple circuit could be created using the following code

```python
from qiskit import QuantumCircuit

# circuit with 1 quibit
simple_circuit = QuantumCircuit(1)
simple_circuit.draw('mpl')
```

The above qubit does nothing

Now let's define a CNOT gate with qubit 0 as control and qubit 1 as target

```python
multi_qubit_circuit = QuantumCircuit(9)
multi_qubit_circuit.cx(0,1)
multi_qubit_circuit.draw('mpl')
```

Acting on state vector `sv_multi`

```python
new_sv_multi = sv_multi.evolve(multi_qubit_circuit)
print(new_sv_multi)
plot_state_qsphere(new_sv_multi.data)
```

The original state is $\vert001001\rangle$. Since qubit 0 is 1 and the qubit 1 is 0, the CNOT gate flips qubit 1 from 0 to 1

### Bell state

The following circuit could form the basis of bell states

```python
bell_circuit = QuantumCircuit(2)
bell_circuit.h(1)
bell_circuit.cx(1, 0)
bell_circuit.draw('mpl')
```

```python
state_zero = Statevector.from_label('00')

bell_state_zero = state_zero.evolve(bell_circuit)
print(bell_state_zero)
plot_state_qsphere(bell_state_zero.data)
```

### Bell Measurement

The following circuit defines a bell measurement

```python
bell_measurement_circuit = QuantumCircuit(2, 2) #This creates a quantum circuit that has 2 qubits and 2 classical bits
bell_measurement_circuit.cx(1, 0)
bell_measurement_circuit.h(1)
bell_measurement_circuit.measure(1, 0)
bell_measurement_circuit.measure(0, 1)
bell_measurement_circuit.draw('mpl')
```

Now we initialize the circuit with $ψ^{01} = (\vert01\rangle+\vert10\rangle)/\sqrt{2}$. Note that we are initializing two qubits and so the first argument in the `initialize` funciton is a 4-element array and the second element indicating it is initializing the first and second index

```python
import numpy as np

bell_measurement_circuit = QuantumCircuit(2, 2) #This creates a quantum circuit that has 2 qubits and 2 classical bits
bell_measurement_circuit.cx(1, 0)
bell_measurement_circuit.h(1)
bell_measurement_circuit.measure(1, 0)
bell_measurement_circuit.measure(0, 1)
bell_measurement_circuit.draw('mpl')

bell_measurement_circuit.initialize([0, np.sqrt(0.5), np.sqrt(0.5), 0], [0, 1])
bell_measurement_circuit.draw('mpl')
```

Now we try with $ψ^{01} = (\vert00\rangle+\vert11\rangle)/\sqrt{2}$




```python
# Initialize state vector example:

test_vector = Statevector([np.sqrt(0.5), 0, 0, np.sqrt(0.5)])
plot_state_qsphere(test_vector.data)

bell_measurement_circuit = QuantumCircuit(2, 2) #This creates a quantum circuit that has 2 qubits and 2 classical bits
bell_measurement_circuit.cx(1, 0)
bell_measurement_circuit.h(1)
bell_measurement_circuit.barrier()
bell_measurement_circuit.measure(1, 0)
bell_measurement_circuit.measure(0, 1)
bell_measurement_circuit.draw('mpl')


bell_measurement_circuit.initialize(test_vector)
bell_measurement_circuit.draw('mpl')
```

### Bell state measurement simulation


Now we want to run a similator on the bell measurement. We will be using the qsam simulator, which produces result counts on differnt measurements. However, for a two quibit circuit, the input is always $\vert00\rangle$. Therefore some gates will need to be applied to convert the states to the one we want

```python
bell_measurement_circuit_zero_zero = QuantumCircuit(2, 2) # |00>
bell_measurement_circuit_zero_zero.h(1)
bell_measurement_circuit_zero_zero.cx(1,0)
bell_measurement_circuit_zero_zero.barrier()

bell_measurement_circuit_zero_zero.cx(1, 0)
bell_measurement_circuit_zero_zero.h(1)
bell_measurement_circuit_zero_zero.measure([0,1],[0,1])
bell_measurement_circuit_zero_zero.draw('mpl')
```

Now let's run the simiulator

```python
from qiskit import transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram

simulator = Aer.get_backend('qasm_simulator')
new_circuit = transpile(bell_measurement_circuit_zero_zero, simulator)
job = simulator.run(new_circuit, shots=10000)
counts = job.result().get_counts(new_circuit)
plot_histogram(counts)
```

All measurement of the qubits are 0, which is expected

Now let's try $\vert01\rangle$

```python
bell_measurement_circuit_zero_one = QuantumCircuit(2, 2) # |00>
bell_measurement_circuit_zero_one.x(0)
bell_measurement_circuit_zero_one.barrier()
bell_measurement_circuit_zero_one.h(1)
bell_measurement_circuit_zero_one.cx(1,0)
bell_measurement_circuit_zero_one.barrier()

bell_measurement_circuit_zero_one.cx(1, 0)
bell_measurement_circuit_zero_one.h(1)
bell_measurement_circuit_zero_one.measure([0,1],[0,1])
bell_measurement_circuit_zero_one.draw('mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
new_circuit = transpile(bell_measurement_circuit_zero_one, simulator)
job = simulator.run(new_circuit, shots=10000)
counts = job.result().get_counts(new_circuit)
plot_histogram(counts)
```

The result is 01, which is expected
